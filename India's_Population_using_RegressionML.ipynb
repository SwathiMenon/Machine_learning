AIM:To predict India's Population

Dataset Link:https://www.kaggle.com/code/dinashboul/linear-regression

containing Indian population data to predict India's Population





```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/datasets/population.csv")
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
#label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["World Population"]=le.fit_transform(df["World Population"])
```

**LINEAR REGRESSION**

```python
x1=df.iloc[:,0:1]
x1
```

```python
y1=df.iloc[:,-1]
y1
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
y_pred=mlr.predict(X_test)
y_pred
```

```python
#MULTIPLE REG PERFORMANCE_MEASURE
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("mean absolute error",mean_absolute_error(y_test,y_pred))
print("mean squred error",mean_squared_error(y_test,y_pred))
print("root mean squred error",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score",r2_score(y_test,y_pred))
```
