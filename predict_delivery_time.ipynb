```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/datasets/porter1.csv")
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
#CONVERTING TO DATE-TIME DATA TYPE
df['actual_delivery_time'] = pd.to_datetime(df['actual_delivery_time'])
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df.drop(["market_id"],inplace=True,axis=1)
df.drop(["store_id"],inplace=True,axis=1)
df.drop(["store_primary_category"],inplace=True,axis=1)
```

```python
df
```

```python
#FINDING THE ACTUVAL DELIVERY TIME 
df['created_at'] = df['actual_delivery_time'] - df['created_at']
#CONVERTING IN TO MINUTES
from datetime import timedelta
df['created_at'] = df['created_at'] /timedelta(minutes=1)
df.drop(["actual_delivery_time"],inplace=True,axis=1)
```

```python
df
```

```python
df['min_item_price'] = (df['min_item_price'] + df['max_item_price']) / 2
df.drop(["max_item_price"],inplace=True,axis=1)
```

```python
df = df.rename(columns={'created_at': 'del_time'})
df = df.rename(columns={'min_item_price': 'avg_amount'})
```

```python
df
```

```python
df.isna().sum()
```

```python
3#removing columns with null value
df = df.dropna()
```

```python
#removing columns with negative value
df = df[df >= 0].dropna()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,1:]
x
```

```python
y=df.iloc[:,1]
y
```

```python
#SCALING
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
scx=mms.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scx,y,random_state=1,test_size=.5)
```

```python
#MULTIPLE REG MODEL BUILDING
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
y_pred=mlr.predict(x_test)
y_pred
```

```python
#MULTIPLE REG PERFORMANCE_MEASURE
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("mean absolute error",mean_absolute_error(y_test,y_pred))
print("mean squred error",mean_squared_error(y_test,y_pred))
print("root mean squred error",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score",r2_score(y_test,y_pred))
```

```python
#SVR MODEL BUILDING
from sklearn.svm import SVR
sv=SVR()
sv.fit(x_train,y_train)
yp=sv.predict(x_test)
yp
```

```python
#SVR PERFORMANCE_MEASURE
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("mean absolute error",mean_absolute_error(y_test,yp))
print("mean squred error",mean_squared_error(y_test,yp))
print("root mean squred error",np.sqrt(mean_squared_error(y_test,yp)))
print("r2_score",r2_score(y_test,yp))
```
